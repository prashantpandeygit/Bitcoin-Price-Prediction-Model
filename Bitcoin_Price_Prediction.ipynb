### ***PRICE PREDICTION USING FEATURES***

```python
X_test
```

```python
def predict_price(lr, open_val, high_val, low_val):
    features = [[open_val, high_val, low_val]]
    predicted_price = lr.predict(features)
    return predicted_price
```

```python
predicted_price = predict_price(lr, open_val=9179.8, high_val=9201.4, low_val=7915.8)
predicted_price
```

### ***PRICE PREDICTION USING DATE***

```python
def predict_price_by_date(lr,date):
    date = pd.to_datetime(date)
    data_for_date = df[df.index == date]
    open_val = data_for_date['open'].values[0]
    high_val = data_for_date['high'].values[0]
    low_val = data_for_date['low'].values[0]
    features = [[open_val, high_val, low_val]]
    predicted_price = lr.predict(features)
    return predicted_price
```

```python
predicted_price = predict_price_by_date(lr,'2021-06-08')
predicted_price
```
